```python
from sklearn.preprocessing import LabelEncoder
import pandas as pd

df = pd.DataFrame({
    'Field': ['Data Science', 'Cybersecurity', 'Web Development', 'AI Engineering']
})

le = LabelEncoder()
df['Field_encoded'] = le.fit_transform(df['Field'])

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Job': ['Data Analyst', 'Software Engineer', 'Network Admin']
})

df_encoded = pd.get_dummies(df, columns=['Job']).astype(int)
print(df_encoded)
```

```python
from sklearn.preprocessing import OrdinalEncoder
import pandas as pd

df = pd.DataFrame({
    'Experience': ['Junior', 'Intermédiaire', 'Senior']
})

encoder = OrdinalEncoder(categories=[['Junior', 'Intermédiaire', 'Senior']])
df['Experience_encoded'] = encoder.fit_transform(df[['Experience']])

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Job': ['Data Scientist', 'Web Developer', 'Cybersecurity Specialist',
            'Data Scientist', 'Web Developer'],
    'Salary': [4500, 3000, 4000, 4800, 3200]
})

target_mean = df.groupby('Job')['Salary'].mean().to_dict()
df['Job_target_encoded'] = df['Job'].map(target_mean)

print(df)
```

```python
import pandas as pd

df = pd.DataFrame({
    'Job': ['Backend Dev', 'Backend Dev', 'DevOps Engineer',
            'Data Engineer', 'DevOps Engineer']
})

freq = df['Job'].value_counts(normalize=True).to_dict()
df['Job_freq_encoded'] = df['Job'].map(freq)

print(df)
```

```python
from tensorflow.keras.layers import Input, Embedding, Flatten
from tensorflow.keras.models import Model
import numpy as np

# Supposons 6 métiers différents, embedding dimension = 4
input_dim = 6
embed_dim = 4

inputs = Input(shape=(1,))
embedding_layer = Embedding(input_dim=input_dim, output_dim=embed_dim)(inputs)
flat = Flatten()(embedding_layer)

model = Model(inputs, flat)
model.summary()

# Exemple d'encodage du métier indexé "3" (ex: DevOps)
encoded = model.predict(np.array([3]))
print(encoded)
```
